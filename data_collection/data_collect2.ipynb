```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.common.keys import Keys
```

```python
service = Service(EdgeChromiumDriverManager().install())
driver = webdriver.Edge(service=service)
driver.maximize_window()

homepage_url = "https://www.bandsintown.com/?came_from=257&utm_source=home&utm_medium=web&utm_campaign=top_event"
collected_concerts = set()
dataa = []



def scrape_concerts( homepage_url):
    """Scrape les concerts pour un genre donn√©."""
    global collected_concerts, dataa
    print(f"üîç Scraping genre...")
    driver.get(homepage_url)
    time.sleep(5)  # Attente du chargement initial
     
    
    
    # Attendre que l'utilisateur change la localisation
    print("‚è≥ Veuillez changer manuellement la localisation √† 'New York City, NY'.")
    print("‚è≥ Vous avez 40 secondes pour effectuer cette op√©ration.")
    for i in range(40, 0, -1):
        if i % 5 == 0:  # Afficher le d√©compte toutes les 5 secondes
            print(f"‚è≥ {i} secondes restantes...")
        time.sleep(1)
    
    print("‚úÖ Reprise du scraping...")
    time.sleep(3)  # Petit d√©lai suppl√©mentaire pour s'assurer que tout est charg√©
    
    # Commencer le scraping
    last_height = driver.execute_script("return document.body.scrollHeight")
    scroll_count = 0
    max_scrolls = 20 
    while scroll_count < max_scrolls:
            concerts = driver.find_elements(By.CSS_SELECTOR, 'a.HsqHp2xM2FkfSdjy1mlU')

            for concert in concerts:
                try:
                    name = concert.find_element(By.CSS_SELECTOR, 'div._5CQoAbgUFZI3p33kRVk').text
                    popularity = [popul.text for popul in concert.find_elements(By.CSS_SELECTOR, 'div.uemLS9AP30BLmy0YGHSY > div')]

                    # Ajouter uniquement les nouveaux concerts
                    if name not in collected_concerts:
                        collected_concerts.add(name)
                        dataa.append([name, ", ".join(popularity)])
                except Exception:
                    continue

            # Scroll vers le bas
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

            # V√©rifier si on est arriv√© en bas
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
scrape_concerts(homepage_url)
filename = "bandsintown_concerts_alteernative.csv"
with open(filename, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'popularity'])
    writer.writerows(dataa)

# Fermer le navigateur
driver.quit()

print(f"‚úÖ Scraping termin√©. {len(dataa)} concerts enregistr√©s dans '{filename}'")

```
