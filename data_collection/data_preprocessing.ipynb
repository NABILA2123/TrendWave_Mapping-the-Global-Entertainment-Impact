```python
df = pd.read_csv('./chartmetric_playlists.csv')

# 2. Nettoyage des données
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)  # ou imputation des valeurs manquantes

# 3. Traitement des colonnes textuelles
df['genres'] = df['genres'].str.lower().str.split(', ')
df['moods'] = df['moods'].str.lower().str.split(', ')

# 4. Normalisation des données
df['name'] = df['name'].str.lower()

# 5. Encodage des variables catégorielles
encoder = OneHotEncoder()  # Pas besoin de spécifier sparse=False
encoded_platform = encoder.fit_transform(df[['platform']])
encoded_platform_df = pd.DataFrame(encoded_platform.toarray(), columns=encoder.get_feature_names_out(['platform']))
df = pd.concat([df, encoded_platform_df], axis=1)

# 6. Création de nouvelles fonctionnalités
def safe_len(x):
    if isinstance(x, list):  # Vérifier si l'élément est une liste
        return len(x)
    return 0  # Retourner 0 si ce n'est pas une liste (par exemple, NaN ou autre)

df['num_genres'] = df['genres'].apply(safe_len)
df['num_moods'] = df['moods'].apply(safe_len)
```
