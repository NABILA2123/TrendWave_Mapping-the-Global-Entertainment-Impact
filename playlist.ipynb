```python
import yt_dlp

# URL de la playlist
PLAYLIST_URL = "https://www.youtube.com/playlist?list=PL4fGSI1pDJn5kI81J1fYWK5eZRl1zJ5kM"  # Remplacez par l'URL de votre playlist

# Options pour yt-dlp
ydl_opts = {
    "quiet": True,  # Désactive les logs inutiles
    "extract_flat": False,  # Pour obtenir des informations détaillées sur chaque vidéo
    "extract_flat_in_playlist": False,  # Pour extraire les informations complètes des vidéos dans la playlist
}

# Fonction pour extraire les données de la playlist
def extract_playlist_data(playlist_url):
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        # Extraction des informations de la playlist
        playlist_info = ydl.extract_info(playlist_url, download=False)

        # Vérification si l'extraction a réussi
        if not playlist_info:
            print("Erreur : Impossible d'extraire les informations de la playlist.")
            return []

        # Extraction des informations pour chaque vidéo de la playlist
        videos_data = []
        for video in playlist_info["entries"]:
            if video:  # Vérifier que les informations de la vidéo sont disponibles
                video_data = {
                    "title": video.get("title"),
                    "language": video.get("language", "Non spécifié"),
                    "url": video.get("url"),  # URL de la vidéo
                    "uploader": video.get("uploader"),
                    "upload_date": video.get("upload_date"),
                    "duration": video.get("duration"),
                    "view_count": video.get("view_count"),
                    "like_count": video.get("like_count"),
                    "dislike_count": video.get("dislike_count"),
                    "comment_count": video.get("comment_count"),
                    "tags": video.get("tags"),
                    "categories": video.get("categories"),
                    "thumbnail": video.get("thumbnail"),
                }
                videos_data.append(video_data)

        return videos_data

# Extraction des données de la playlist
playlist_data = extract_playlist_data(PLAYLIST_URL)

# Affichage des données extraites
for index, video in enumerate(playlist_data, start=1):
    print(f"Vidéo #{index}:")
    for key, value in video.items():
        print(f"  {key}: {value}")
    print("-" * 40)
```

```python
#transformation du playlist_data en csv
import csv

# Chemin du fichier CSV

CSV_FILE = "playlist_data.csv"

# En-têtes du fichier CSV
CSV_HEADERS = [
    "title",
    "language",
    "url",
    "uploader",
    "upload_date",
    "duration",
    "view_count",
    "like_count",
    "dislike_count",
    "comment_count",
    "tags",
    "categories",
    "thumbnail",
]

# Écriture des données dans le fichier CSV
with open(CSV_FILE, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=CSV_HEADERS)
    writer.writeheader()
    writer.writerows(playlist_data)

print(f"Données de la playlist enregistrées dans le fichier CSV : {CSV_FILE}")

```
